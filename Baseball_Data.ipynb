```python
team_history_url = "https://www.baseball-reference.com/teams/"
```

```python
#read table from url
tables = pd.read_html(team_history_url)
team_history_db = tables[0]
team_history_db
```

```python
team_history_db = team_history_db.loc[team_history_db.Rk.notnull()]
team_history_db
```

```python
team_history_db.sort_values(by=['W-L%'], ascending=False)
```
